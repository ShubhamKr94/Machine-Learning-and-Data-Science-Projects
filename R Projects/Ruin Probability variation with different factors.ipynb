## Ruin Probability vs. Surplus Capital

```
%r
ru = numeric(100)
ini = 1000
ruin_per = c()

for (k in 1:10) {
  
  ini = ini+ 500
  ini
  for (i in 1:1000) {
    
    toc = rexp(100,1)
    aoc = rgamma(100,2,2/1000)
    
    
    annual_pre = 1*mean(aoc)*1.1
    s = numeric(100)
    
    
    for (j in 1:100) {
      if (aoc[j] < 1200){
        
        
        if (j == 1) {
          
          s[j] = ini+annual_pre*toc[j] - aoc[j]
          
          
          
        }
        else {
          s[j] = s[j-1]+annual_pre*toc[j] - aoc[j]
        }
      }
      else
      {
        if (j == 1) {
          
          s[j] = ini+annual_pre*toc[j] - 2000
        }
        else {
          s[j] = s[j-1]+annual_pre*toc[j] - 2000
        }
      }
      
      
    }
    
    
    if (min(s)<0){
      ru[i] = 1
    }
    else{
      ru[i] = 0
    }
    
    
  }
  ruin_per[k]=sum(ru)/length(ru)

}
ruin_per
initial <- seq(1000, 5500, length.out= 10)
initial
plot(initial, ruin_per,type="o", col = "red", xlab = "Surplus", ylab = "Ruin_Probability")
title(main = "Ruin Probability vs. Surplus Capital")
```

## Ruin probability Vs Alpha

```
%r
# Initialize variables
ini <- 1000
ru <- numeric(1000)
ruin_per <- numeric(10)
alpha <- 0

# Loop over different values of alpha
for (k in 1:10) {
  alpha <- alpha + 0.5
  cat("alpha =", alpha, "\n")
  
  for (i in 1:1000) {
    toc <- rexp(100, 1)
    aoc <- rgamma(100, alpha, alpha / 1000)
    annual_pre <- 1 * mean(aoc) * 1.1
    s <- numeric(100)
    
    for (j in 1:100) {
      if (j == 1) {
        s[j] <- ini + annual_pre * toc[j] - aoc[j]
      } else {
        s[j] <- s[j - 1] + annual_pre * toc[j] - aoc[j]
      }
    }
    
    if (min(s) < 0) {
      ru[i] <- 1
    } else {
      ru[i] <- 0
    }
  }
  
  ruin_per[k] <- sum(ru) / length(ru)
}

# Generate alpha values
alpha_values <- seq(0.5, 5, length.out = 10)

# Plotting
plot(alpha_values, ruin_per, type = "o", col = "red", xlab = "alpha", ylab = "Ruin_Probability")
title(main = "Ruin Probability vs. alpha")
grid()
```
